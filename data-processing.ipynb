### Categorical Encoding

To simplify the categorical features in the dataset, we employ categorical encoding for the AccountType, ContentType, and TweetType features. Thus, we encoded these features into their respective numerical values.

For **AccountType**, the encoding was:

- 0 for ‚ÄúAnonymous‚Äù
- 1 for ‚ÄúIdentified‚Äù
- 2 for ‚ÄúMedia‚Äù

and the encoded value was saved as `Temp_AccountType` feature in the `df` dataset.

For **ContentType**, the encoding was:

- 0 for ‚ÄúEmotional‚Äù
- 1 for ‚ÄúRational‚Äù, and
- 2 for ‚ÄúTransactional‚Äù

and the encoded value is saved as `Temp_ContentType` feature in the `df` dataset.

For the TweetType, we split the string of the feature to tokenize the content of the column into a list. Then, we mapped the list of tokens of the TweetType column via <i>one hot encoding</i> to see if the tweet contains certain tweet_types (Text, Image, URL, and Reply) by putting a value of 1 to their respective feature under the `df` dataset.

```python
from sklearn import preprocessing

encoder_account_type =  preprocessing.LabelEncoder().fit(df['AccountType'])
df['Temp_AccountType'] = encoder_account_type.transform(df['AccountType'])

encoder_content_type =  preprocessing.LabelEncoder().fit(df['ContentType'])
df['Temp_ContentType'] = encoder_content_type.transform(df['ContentType'])
```

```python
# Onehot encoding for TweetType
def tweet_type_split(x):
  return x.split(', ')

def tweet_type_classification(tweettype, x):
  if tweettype in x:
    return 1
  return 0

df['TweetType'] = df['TweetType'].apply(tweet_type_split)

tweet_types = ['Text', 'Image', 'URL', 'Reply']
for tweet_type in tweet_types:
  df[f'Temp_TweetType{tweet_type}'] = df['TweetType'].apply(lambda x: tweet_type_classification(tweet_type, x))
```

###Binning Date Features


In order to organize the continuous date and time data, it was divided into either **Monthly** or **Yearly** bins or intervals. 

For the **DatePosted** Feature, the datetime data type is converted into periods using the `to_period(‚Äôm‚Äô)` function. Hence, the DatePosted feature now contains the month when the tweet was posted. Then, the column was divided into monthly interval bins by using `post_bins = pd.period_range(post_start, post_end, freq='M')` where `post_start` is the earliest DatePosted month into the dataset while the `post_end` is the latest and the `freq` or frequency is set to **‚ÄòM‚Äô** or **monthly**. Thus, the post_bins contains the binning of the DatePosted feature returned by the period_range function which returns a fixed monthly frequency PeriodIndex.

The JoinedDate feature is converted into yearly interval using the `PeriodIndex(df[‚ÄôJoined‚Äô]).year`. Thus, the added temporary feature in the dataset **Temp_YearJoined** now contains the year when the account was created in twitter.

```python
# Binning DatePosted feature into months
df['Temp_MonthPosted'] = pd.to_datetime(df['DatePosted']).dt.to_period('m').astype('str')

post_start = df['Temp_MonthPosted'].min()
post_end = df['Temp_MonthPosted'].max()

post_bins = pd.period_range(post_start, post_end, freq='M')

# Binning JoinedDate into years
df['Temp_YearJoined'] = pd.PeriodIndex(df['Joined']).year
```

### Interpolation

  To aggregate and interpolate the data, we first created an empty df dataset named `df_time_series` in which the features and rows are characterized with:
  - First feature is the total count of tweets per time bin
  - The succeeding features are each type of content and account separated by each column, and
  - The rows are the indices of the time binning with monthly frequency

  Then, the `count_time_series` function counts the number of tweets for each column depending on the content of the tweet and account type of the user of the tweet. Note that there will be null values in `df_time_series` since the values of the columns in `df_time_series` will only increase if it is encountered in the dataset.

  Then, we interpolated the `df_time_series` using the interpolate function with **linear** method. Moreover, in order to remove the leading NaN values in the dataset, we filled the values with 0 using `fillna(0)` method.



```python
# CHANGED

import numpy as np

ContentType = ['ContentType_Rational', 'ContentType_Emotional', 'ContentType_Transactional']
AccountType = ['AccountType_Anonymous', 'AccountType_Identified', 'AccountType_Media']

df_time_series = pd.DataFrame(columns=['TotalCount']+ContentType+AccountType, index=post_bins)
df_time_series.index.name="MonthPosted"

for col in df_time_series.columns:
    df_time_series[col] = df_time_series[col].astype(float)

def count_time_series(x,df_op):
  content = encoder_content_type.inverse_transform([x['Temp_ContentType']])[0]
  account = encoder_account_type.inverse_transform([x['Temp_AccountType']])[0]
  if np.isnan(df_op['TotalCount'].loc[x['Temp_MonthPosted']])==True:
    df_op['TotalCount'].loc[x['Temp_MonthPosted']]=0
  if np.isnan(df_op[f'ContentType_{content}'].loc[x['Temp_MonthPosted']])==True:
    df_op[f'ContentType_{content}'].loc[x['Temp_MonthPosted']]=0
  if np.isnan(df_op[f'AccountType_{account}'].loc[x['Temp_MonthPosted']])==True:
    df_op[f'AccountType_{account}'].loc[x['Temp_MonthPosted']]=0

  df_op['TotalCount'].loc[x['Temp_MonthPosted']] = df_op[f'TotalCount'].loc[x['Temp_MonthPosted']] +1
  df_op[f'ContentType_{content}'].loc[x['Temp_MonthPosted']] = df_op[f'ContentType_{content}'].loc[x['Temp_MonthPosted']] +1
  df_op[f'AccountType_{account}'].loc[x['Temp_MonthPosted']] = df_op[f'AccountType_{account}'].loc[x['Temp_MonthPosted']] +1
      
df.apply(lambda x: count_time_series(x,df_time_series), axis=1)

df_time_series = df_time_series.interpolate(method='linear').fillna(0)
df_time_series
```

### Natural Language Processing

In order to obtain a proper analysis of the tweets in our dataset, the removal of stop words, links, mentions, hashtags, and some punctuations in the tweets were performed on the contents of each tweet.

These removals are necessary in order to keep the focus on key points of the tweets.

```python
# Remove stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string

stop_words = stopwords.words('english')
f=open('/content/stopwords.txt')
filipino_stopwords = f.read().splitlines()
stop_words+=filipino_stopwords+[*string.punctuation]

df['Tweet_stop'] = df['Tweet'].str.lower()
df['Tweet_stop'] = df['Tweet_stop'].apply(lambda words: ' '.join(word for word in words.split() if word not in stop_words))
df.head(3).style
```

```python
# removes links, mentions, hashtags, and some punctuations in the tweets
import re

def text_extract(tweet):
    http_link = re.sub(r'\shttp\S*', '', tweet)
    com_link = re.sub(r'\s\S+.com\S*', '', http_link)
    mentions = re.sub(r'@\S+\s*','',com_link)
    alt_text = re.sub("[\{].*?[\}]", "", mentions)
    hashtags = re.sub(r'#\S+\s*','',alt_text)
    return hashtags

def nreplace(tweet):
    return tweet.replace('\n',' ')

def lowercase(tweet):
    return tweet.lower()
punctuations = ''.join(['!', '"', '#', '$', '%',  "'", '(', ')', '*', '+', ',', '.', ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`', '{', '|', '}', '~'])
df['Tweet_stop'] = df['Tweet_stop'].map(text_extract).map(nreplace).map(lowercase).apply(lambda words: words.translate(str.maketrans('', '', punctuations)))
df[['Tweet','Tweet_stop']].head(5)
```

After removing the unnecessary parts of the tweets, we proceed with obtaining the top 10 most frequent words used in the tweets.

```python
# Count word frequency
print("Most commonly used words in tweets:")
word_freq = df['Tweet_stop'].str.lower().str.split().explode().value_counts()
word_freq[:10].to_frame().style
```
